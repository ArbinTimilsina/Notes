# Matplotlib
**This note uses Python 2.7**

**Matplotlib** is a python 2D plotting library which produces publication quality figures.

### Single plot

```python
%matplotlib inline
import matplotlib.pyplot as plt

#Generage data
import numpy as np
np.random.seed(0)
myData1 = np.random.normal(-50, 50, 10000)
myData2 = np.random.normal(0, 25, 10000)

myFigure = plt.figure(figsize=(9, 6), dpi=80, facecolor='w', edgecolor='k') 
#figsize: w,h tuple in inches, dpi= dots per inch

myAxis = myFigure.add_subplot(1,1,1) 
#The figure should be 1x1 and we are selecting the first subplot

myAxis.hist(myData1, 100, range = (-200, 200), color='g', alpha=0.5, label = 'My data one')
myAxis.hist(myData2, 100, range = (-200, 200), color='b', alpha=0.5, label = 'My data two')

#Title and stuff
myAxis.set_title('My Plot', fontsize=25)
myAxis.set_xlabel('x-axis (units)', fontsize=20)
myAxis.set_ylabel('y-axis (units)', fontsize=20)

#Legend
myAxis.legend(loc='best')

#Ticks
for tick in myAxis.xaxis.get_ticklabels():
    tick.set_fontsize('large')
    #ick.set_fontname('Times New Roman')
    #tick.set_color('blue')
    tick.set_weight('bold')
    
for tick in myAxis.yaxis.get_ticklabels():
    tick.set_fontsize('large')
    #tick.set_fontname('Times New Roman')
    #tick.set_color('blue')
    tick.set_weight('bold')    
```

### Multiple plots

```python
def setTitle(axis, title, xAxis, yAxis):
    axis.set_title(title, fontsize=25)
    axis.set_xlabel(xAxis, fontsize=20)
    axis.set_ylabel(yAxis, fontsize=20)

def setTicks(axis):    
    for tick in axis.xaxis.get_ticklabels():
        tick.set_fontsize('large')
        tick.set_weight('bold')
    
    for tick in axis.yaxis.get_ticklabels():
        tick.set_fontsize('large')
        tick.set_weight('bold')
    
myFigure2 = plt.figure(figsize=(18, 12), dpi=80, facecolor='w', edgecolor='k') 
myAxis1 = myFigure2.add_subplot(2,2,1)
myAxis2 = myFigure2.add_subplot(2,2,2)

myAxis1.hist(myData1, 100, range = (-200, 200), color='g', alpha=0.5, label = 'My data one')
myAxis2.hist(myData2, 100, range = (-200, 200), color='b', alpha=0.5, label = 'My data two')

#Title and stuff
setTitle(myAxis1, 'My Plot one', 'x-axis (units)', 'y-axis (units)')
setTitle(myAxis2, 'My Plot two', 'x-axis (units)', 'y-axis (units)')

#Legend
myAxis1.legend(loc='best')
myAxis2.legend(loc='best')

#Ticks
setTicks(myAxis1)
setTicks(myAxis2)
```
